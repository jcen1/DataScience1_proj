```python
import gzip
import pandas as pd
with gzip.open('daily_dataset.csv.gz') as f:

    daily_data = pd.read_csv(f, parse_dates = ["day"], index_col = "day")

daily_data.head()
```

```python
daily_data.describe()
```

```python
daily_data.isnull().sum()
```

```python
#fill in 30 missing values based on their LCLid
 
daily_data['energy_sum'].fillna(daily_data.groupby("LCLid")["energy_sum"].transform('mean'), inplace = True)
```

```python
#check missing dates 
print(pd.date_range(
  start=daily_data.index.min(), end=daily_data.index.max()).difference(daily_data.index))
```

```python
daily_data.groupby(['day']).size().plot(figsize=(25,5))
```

# get weather data

```python
weather = pd.read_csv('weather_daily_darksky.csv')
weather.head().transpose()
```

```python
weather.describe().transpose()
```

```python
#no missing values
weather.isnull().sum()
```

```python
weather['day']=  pd.to_datetime(weather['time']) # day is given as timestamp
weather['day']=  pd.to_datetime(weather['day'],format='%Y%m%d')
weather

#need to compare min and max to make sure everything has covered 
```

```python
#check missing dates
weather = weather.set_index('day')
#https://stackoverflow.com/a/52044522
#print(pd.date_range(start='2011-09-01', end='2011-10-25').difference(weather.index))
print(weather.reindex(pd.date_range(weather.index.min(), weather.index.max())).isnull().all(1))
print("weather min day is {}, electric min day is {}".format(weather.index.min(),daily_household.day.min()))
print("weather max day is {}, electric max day is {}".format(weather.index.max(),daily_household.day.max()))
```

```python
weather['month'] =pd.DatetimeIndex(weather.index).month
weather['uvIndex'].fillna(weather.groupby("month")["uvIndex"].transform('mean'), inplace = True)
```

# Feature Eng

```python
#add average temp 

weather['avg_temp'] = weather[['temperatureMax', 'temperatureMin']].mean(axis=1)
weather['apparent_avg_temp'] = weather[['apparentTemperatureHigh', 'apparentTemperatureLow']].mean(axis=1)
```

```python
#apply season 
# https://stackoverflow.com/a/63640283
# https://stackoverflow.com/a/44124490
weather['season']= pd.DatetimeIndex(weather.index).month%12 // 3 + 1
weather.head().transpose()
```

```python
#merge files
weather = weather.reset_index()
weather_merge= weather[[ 'windBearing', 'dewPoint', 'cloudCover', 'windSpeed',
       'pressure',  'visibility', 'humidity',
       'uvIndex','moonPhase','day','avg_temp','apparent_avg_temp','season']]
daily_household_temp = daily_household.reset_index(drop = True).merge(weather_merge,on='day',how='left').set_index('day')
daily_household_temp
```
